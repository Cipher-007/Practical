```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Airquality.csv")
heart = pd.read_csv("Heart.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Ozone"] = df["Ozone"].fillna(df["Ozone"].mean())
df["Solar.R"] = df["Solar.R"].fillna(df["Solar.R"].mean())
```

```python
df.isnull().sum()
```

```python
df.head()
```

SCATTER PLOT

```python
plt.scatter(x = df["Temp"], y = df["Ozone"], color = "red")
plt.title("Temperature")
plt.xlabel("Temp")
plt.ylabel("Ozone")
plt.show()
```

PIE CHART

```python
data = [df["Ozone"].mean(),df["Wind"].mean(),df["Temp"].mean(),df["Solar.R"].mean()]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.1, 0.1, 0.1)
labels_name = ["Ozone","Wind","Temp","Solar.R"]
plt.pie(data,explode=explode,labels = labels_name, autopct='%1.1f%%', colors=colors,shadow=True,)
```

Bar Graph

```python
fig = plt.figure(figsize =(10, 7))
plt.bar(labels_name, data)
```

HISTOGRAM

```python
sns.distplot(df['Temp'])
```

LINE GRAPH

```python
h = df.iloc[1:16,1]
v = df.iloc[1:16,3]
plt.plot(h,label = 'ozone',marker = 'o',linestyle = "dotted")
plt.plot(v,label = 'wind',marker = 'o',linestyle = "dashed")
plt.ylim(0,60)
plt.legend()
plt.title("Ozone and Wind chart for 1st-15th May")
plt.ylabel("Value Range")
plt.xlabel("Days")
```

PAIR PLOT

```python
sns.set(style = "ticks")
sns.pairplot(df)
plt.show()
```
